Short explanation of the describe measures:
- count: Indication of how many values are present in the columns (NaNs/missing values are not counted).
- mean: average value of the data
- std: standard deviation of the data
- min: the smallest value in the data set
- 25%: 25 % of the data are below this value
- 50%: 50% of the data are below this value. This value is called the median.
- 75%: 75% of the data are below this value
- max: the largest expression in the data set

```python
import matplotlib.pyplot as plt

# Define the list of federal states in Germany
federal_states = ['Baden-Württemberg', 'Bavaria', 'Berlin', 'Brandenburg', 'Bremen', 'Hamburg', 'Hesse',
                  'Lower Saxony', 'Mecklenburg-Vorpommern', 'North Rhine-Westphalia', 'Rhineland-Palatinate',
                  'Saarland', 'Saxony', 'Saxony-Anhalt', 'Schleswig-Holstein', 'Thuringia']

# Extract the relevant data for grades until 2019
grades_until_2019 = avg_noten_concat[avg_noten_concat['year'] <= '2019-01-01']

# Calculate the average grade for each federal state until 2019
avg_grades_until_2019 = grades_until_2019.groupby('year').mean()

# Extract the relevant data for grades from 2020 till 2022
grades_from_2020_to_2022 = avg_noten_concat[avg_noten_concat['year'] >= '2020-01-01']

# Calculate the average grade for each federal state from 2020 till 2022
avg_grades_from_2020_to_2022 = grades_from_2020_to_2022.groupby('year').mean()

# Plot the average grades until 2019
plt.figure(figsize=(10, 6))
for state in federal_states:
    plt.plot(avg_grades_until_2019.index, avg_grades_until_2019[state], label=state)
plt.title('Average Grades Until 2019')
plt.xlabel('Year')
plt.ylabel('Average Grade')
plt.legend(loc='best')
plt.show()

# Plot the average grades from 2020 till 2022
plt.figure(figsize=(10, 6))
for state in federal_states:
    plt.plot(avg_grades_from_2020_to_2022.index, avg_grades_from_2020_to_2022[state], label=state)
plt.title('Average Grades From 2020 to 2022')
plt.xlabel('Year')
plt.ylabel('Average Grade')
plt.legend(loc='best')
plt.show()
```
