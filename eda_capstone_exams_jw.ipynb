Short explanation of the describe measures:
- count: Indication of how many values are present in the columns (NaNs/missing values are not counted).
- mean: average value of the data
- std: standard deviation of the data
- min: the smallest value in the data set
- 25%: 25 % of the data are below this value
- 50%: 50% of the data are below this value. This value is called the median.
- 75%: 75% of the data are below this value
- max: the largest expression in the data set

```python
import matplotlib.pyplot as plt
import numpy as np

# generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create a figure and axis objects
fig, ax = plt.subplots()

# plot the data
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# find the maximum and minimum values for both y1 and y2
y1_max = max(y1)
y1_min = min(y1)
y2_max = max(y2)
y2_min = min(y2)

# set the fontsize of the table text
fontsize = 10

# create a table with the maximum and minimum values for y1 and y2
table_text = [['', 'sin(x)', 'cos(x)'],
              ['Max', f'{y1_max:.2f}', f'{y2_max:.2f}'],
              ['Min', f'{y1_min:.2f}', f'{y2_min:.2f}']]

# add the table to the plot
ax.table(cellText=table_text, loc='lower center', fontsize=fontsize)

# set the legend
ax.legend()

# display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create the plot
fig, ax = plt.subplots()

# plot the data
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# calculate and plot the max and min values for each line
y1_max, y1_min = np.max(y1), np.min(y1)
y2_max, y2_min = np.max(y2), np.min(y2)

ax.plot([np.argmax(y1), np.argmax(y1)], [y1_max, y1_max+0.2], 'r', label=f'max={y1_max:.2f}')
ax.plot([np.argmin(y1), np.argmin(y1)], [y1_min, y1_min-0.2], 'r', label=f'min={y1_min:.2f}')

ax.plot([np.argmax(y2), np.argmax(y2)], [y2_max, y2_max+0.2], 'g', label=f'max={y2_max:.2f}')
ax.plot([np.argmin(y2), np.argmin(y2)], [y2_min, y2_min-0.2], 'g', label=f'min={y2_min:.2f}')

# set the legend location and add a title
ax.legend(loc='upper right')
ax.set_title('Sine and Cosine')

plt.show()
```
