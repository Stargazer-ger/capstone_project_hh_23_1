```python
# Group states by regions
east_states = ['BB', 'BE', 'MV', 'SN', 'ST', 'TH']
west_states = ['NW', 'RP', 'HE', 'SH', 'NI']
north_states = ['HH', 'NI', 'SH', 'MV']
south_states = ['BY', 'BW', 'HE', 'RP', 'SL']

# Create a dictionary to map states to regions
state_region_dict = {}
for state in noten_2022.columns[0:]:
    if state in east_states:
        state_region_dict[state] = 'East'
    elif state in west_states:
        state_region_dict[state] = 'West'
    elif state in north_states:
        state_region_dict[state] = 'North'
    elif state in south_states:
        state_region_dict[state] = 'South'

# Add a new column to the dataframe to represent regions
noten_2022['Region'] = noten_2022.columns[0:].map(state_region_dict)

# Group by region and sum the values
#df_grouped = noten_2022.groupby('Region').sum()

# Reset the index of the grouped dataframe
#df_grouped = df_grouped.reset_index()

# Print the grouped dataframe
#print(df_grouped)
```

```python
# Calculate the percentage change in number of students by state
percent_change_by_state = noten_2022.pct_change(axis=0) * 100
print(percent_change_by_state)
```

Short explanation of the describe measures:
- count: Indication of how many values are present in the columns (NaNs/missing values are not counted).
- mean: average value of the data
- std: standard deviation of the data
- min: the smallest value in the data set
- 25%: 25 % of the data are below this value
- 50%: 50% of the data are below this value. This value is called the median.
- 75%: 75% of the data are below this value
- max: the largest expression in the data set
