```python
import pandas as pd
import matplotlib.pyplot as plt
from master_NZ import load_students_per_teacher_data

# Erster Balken
def load_all_students_per_teacher_data():
    data = {}
    for year in range(2011, 2022):
        data[year] = load_students_per_teacher_data(year)
    return data

students_per_teacher_data = load_all_students_per_teacher_data()

#students_per_teacher = load_students_per_teacher_data(2011)
#students_per_teacher = students_per_teacher[students_per_teacher['SchoolType'] == 'Allgemein bildende Schulen']
#students_per_teacher = students_per_teacher[['Year', 'Deutschland']]
#students_per_teacher = students_per_teacher.set_index('Year')

# Zweiter Balken
graduation_rates = {'Federal States': ['Hamburg', 'Berlin', 'Brandenburg', 'Nordrhein-Westfalen', 'Bremen',
                                       'Mecklenburg-Vorpommern', 'Rheinland-Pfalz', 'Schleswig-Holstein', 'Saarland',
                                       'Thüringen', 'Hessen', 'Sachsen', 'Niedersachsen', 'Baden-Württemberg', 'Bayern',
                                       'Sachsen-Anhalt'],
                    'avg_abitur_graduation_rate': [54.24, 43.28, 40.18, 38.59, 37.33, 34.35, 33.98, 33.97, 33.65,
                                                  32.85, 32.75, 30.95, 30.60, 30.09, 28.90, 28.26]}
graduation_rates_df = pd.DataFrame(graduation_rates)
# Zusammenführen der Daten
data = students_per_teacher.join(graduation_rates)

# Säulendiagramm zeichnen
fig, ax = plt.subplots()
data.plot(kind='bar', ax=ax)
ax.set_title('Daten zur Visualisierung')
ax.set_xlabel('Bundesländer')
ax.set_ylabel('Anzahl der Schüler pro Lehrer und Abiturabschlussquote')
plt.show()
```
