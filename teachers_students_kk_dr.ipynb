```python
# Filter by school types
school_type = ['Allgemeinbildende Schulen']
allgemeinbildende_schulen = students_per_teacher_concat[students_per_teacher_concat['SchoolType'].isin(school_type)]
allgemeinbildende_schulen
```

```python
avg_students_per_teacher = allgemeinbildende_schulen.groupby('Year').sum().mean().reset_index()
avg_students_per_teacher.rename(columns={'index':'Federal States',
                                        0: 'avg_students_per_teacher'}, inplace=True)
```

```python
avg_students_per_teacher
```

```python
# drop Deutschland
avg_students_per_teacher_without_germany = avg_students_per_teacher.iloc[:16]
```

```python
# Sort the states based on average Pass Rate
# top_states_highest = avg_students_per_teacher.sort_values('avg_students_per_teacher', ascending=False).head(16)
top_states_lowest = avg_students_per_teacher_without_germany.sort_values('avg_students_per_teacher', ascending=True).head(16)

# plt.figure(figsize=(13, 7))
# #plt.rcParams.update({'font.size': 14})
# # Create a bar chart for top 3 states with highest Pass Rate
# plt.bar(top_states_highest['Federal States'], top_states_highest['avg_students_per_teacher'], color='#002845')
# plt.grid(False)  # Add grid lines
# plt.tick_params(axis='x')#, rotation=75)
# #plt.xlabel('State')
# plt.ylabel('Highest Students-Teacher Ratio')
# #plt.title('Top 5 States with Highest Passing Rate in the Last Decade')
# plt.ylim(bottom=11)  # Set the y-axis lower limit to 0

# # Add value labels on top of each bar
# for i, v in enumerate(top_states_highest['avg_students_per_teacher']):
#     plt.text(i, v + 0.1, str(round(v, 2)), ha='center', color='black')
# plt.show()

farben = ['darkturquoise', '#002845', '#002845', 'darkturquoise', 'darkturquoise', '#002845', '#002845', '#002845', '#002845', '#002845',
           'darkturquoise', '#002845', '#002845', '#002845', 'lightcoral', '#002845']

plt.figure(figsize=(13, 7))
# Create a bar chart for top 3 states with lowest Pass Rate
plt.bar(top_states_lowest['Federal States'], top_states_lowest['avg_students_per_teacher'], color=farben)

plt.grid(False)  # Add grid lines
plt.tick_params(axis='x', rotation=75)
#plt.xlabel('State')
plt.ylabel('Student to teacher ratio')
#plt.title('Top 5 States with Highest Failing Rate in the Last Decade')
plt.ylim(bottom=11)  # Set the y-axis lower limit to 0

# Add value labels on top of each bar
for i, v in enumerate(top_states_lowest['avg_students_per_teacher']):
    plt.text(i, v + 0.05, str(round(v, 2)), ha='center', color='black')

plt.show()
```
